# historic details

#dev used to be #ben-bot-testing, then #ben, then generalized to #dev.

#wiki used to be #hackweek, and was used by ben, ethan, josh, and tristan to make a bot for Discord's Hackweek. They did not end up making a bot. After changing to #wiki, the @majors attempted to make a wiki of our past for fun. They did not end up finishing the wiki.

#gamers used to be #minecraft. Then the server died down, so we expanded the chat to include all gaming (gamers and games).

Before #do was made, _everyone_ was in #dont. After the creation of #do, though, @SilentKON#2259 requested removal. Since he couldn't be let back into #dont, we compromised by making #maybe, which hosted just the majors and @SilenKON#2259.

For some reason, there were only 3 messages sent in #do in all of 2016. There was a **1 year gap** between the third message in #do and the fourth.

Our music bot used to be @God#9608, a bot running locally on Kian's computer. Even back then, though, we had @LewdBot as a backup for when his computer was off.
